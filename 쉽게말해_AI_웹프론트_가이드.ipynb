## 1. 프론트엔드 핵심 기능 목록 및 구현 요구사항

### 1️⃣ 입력 및 업로드 기능 (Input & Upload)
- **텍스트 직접 입력**: 문자메시지, 카카오톡 알림톡 등에서 복사한 복잡한 문장을 바로 붙여넣을 수 있는 텍스트 영역(TextArea) 제공
- **사진/PDF 파일 업로드**: 드래그앤드롭(Drag & Drop) 및 파일 탐색기를 통한 문서 이미지(JPG, PNG) 및 PDF 파일 업로드 UI 구현
- **음성 입력(마이크 버튼)**: Web Speech API(`SpeechRecognition`) 혹은 STT API 연동을 활용하여, 마이크 클릭 시 음성을 인식하고 인식 결과를 입력 창에 텍스트로 보완 자동 입력 (실패 시 일반 텍스트 입력으로 Fallback)
- **개인정보 취급 안내 UI**: 문서를 전송하기 전 개인정보 보호 관련 마스킹 및 안내 문구 명시

### 2️⃣ 분석 상태 피드백 기능 (Processing Status)
- **단계별 진행 상태 표시**: 이미지/텍스트 업로드 후 백엔드 분석 시간 동안 사용자가 이탈하지 않도록 애니메이션 스피너와 현재 진행 상태를 렌더링
  * 상태 구분: `이미지 업로드 중 ➡️ OCR 텍스트 분석 중 ➡️ 개인정보 마스킹 중 ➡️ AI 쉬운 말 요약 및 구조화 중`

### 3️⃣ 구조화된 결과 카드 렌더링 (Result Card UI)
백엔드로부터 수신한 JSON 데이터를 기반으로 가독성이 극대화된 반응형 카드들을 배치 순서(시각적 중요도)에 맞춰 순차적으로 렌더링
- **배치 순서 및 구성 요소**:
  1. **문서 유형 태그**: (예: `예비군 훈련 통지`, `세금 납부 고지` 등) 카테고리 뱃지 표시
  2. **한 줄 결론 카드**: "이 문서는 결국 무엇을 하라는 뜻인가?"를 쉬운 문장으로 직관적으로 노출
  3. **행동 체크리스트 (To-Do)**: 사용자가 실제 수행해야 하는 행동을 단계별 체크박스 리스트로 구성 (인터랙티브 토글 지원)
  4. **중요 날짜 카드**: 마감일, 시작일, 방문일 등을 시각적으로 강조
  5. **준비물 카드**: 신분증, 지참 서류 등 필요 물품 리스트
  6. **주의사항 카드**: 미납 시 가산금, 불참 시 불이익 등 위험 요소를 경고 스타일(붉은색 톤 또는 하이라이트)로 렌더링
  7. **확인 필요 카드**: 원문에 모호하게 쓰여있거나 AI가 확실히 해석하기 어려운 정보(예: 상세 제출처 미비 등)를 별도 확인 필요 사항으로 분리
  8. **원문 근거 매핑**: 요약된 정보의 신뢰도를 높이기 위해, 각 카드 하단에 해당 내용이 원문의 어느 문장에서 발췌되었는지 투명하게 매핑 표시

### 4️⃣ 캘린더 등록 및 알림 연동 기능 (Calendar Integration)
- **사용자 확인 및 수정 모달 (중요)**: AI의 오해석으로 인한 오등록 방지를 위해, 캘린더 등록 버튼 클릭 시 일정 제목, 날짜, 시간, 알림 설정(당일, 1일 전, 7일 전 등)을 사용자가 최종 검토하고 직접 수정할 수 있는 모달 화면 제공
- **구글 캘린더 연동**: 조작 완료 시 Google Calendar 웹의 일정 만들기 템플릿 URL을 생성하여 새 탭 오픈 연동
- **아웃룩 캘린더 연동**: Outlook 라이브 캘린더 템플릿 URL 링크 생성
- **표준 `.ics` 파일 다운로드**: 모바일 및 로컬 아웃룩 등 다양한 캘린더 앱에 즉시 바인딩이 가능하도록 `.ics` 파일의 바이너리 다운로드 트리거 구현

### 5️⃣ 고령자 및 디지털 약자 접근성 기능 (Accessibility Mode)
- **큰 글씨 모드**: 원클릭 토글 버튼을 통해 전체 텍스트 크기를 즉각적으로 키우는 UI 모드 (`text-base` -> `text-2xl` 등 스타일 유연 변환)
- **고대비 테마**: 배경색과 전경색의 대조를 최대로 높여 시력이 떨어지는 사용자가 텍스트를 명확히 구분할 수 있도록 돕는 다크/고대비 전환 토글 지원
- **음성 읽기 기능(TTS)**: 결과 화면에서 `읽어줘` 버튼 클릭 시, Web Speech API의 `SpeechSynthesis`를 통해 한 줄 결론과 해야 할 일을 또렷한 한국어 음성으로 출력

### 6️⃣ 카카오톡 보호자 공유 연동 기능 (KakaoTalk Sharing)
- **카카오톡 공유 API**: 혼자서 행정 처리가 힘든 고령자 사용자가 결과 카드 내용을 복사할 필요 없이, 원클릭으로 자녀나 보호자에게 카카오톡 메시지로 연동된 분석 요약 URL 및 내용을 발송할 수 있도록 연동 및 테스트

## 4. 캘린더 연동 프론트엔드 핵심 로직 시뮬레이션

모달에서 수정 완료된 최종 일정 데이터를 바탕으로 구글 캘린더 URL 및 `.ics` 파일 데이터 블록을 만드는 자바스크립트 변환용 파이썬 구현체입니다.

````python
import urllib.parse
from datetime import datetime, timedelta

def get_google_calendar_url(title, date_str, time_str, description):
    """
    구글 캘린더 등록 외부 링크 파라미터 생성
    """
    start_dt = datetime.strptime(f"{date_str} {time_str}", "%Y-%m-%d %H:%M")
    end_dt = start_dt + timedelta(minutes=30)  # 마감 일정은 30분 기본 설정
    
    start_formatted = start_dt.strftime("%Y%m%dT%H%M%S")
    end_formatted = end_dt.strftime("%Y%m%dT%H%M%S")
    
    base_url = "https://calendar.google.com/calendar/render?action=TEMPLATE"
    params = {
        "text": title,
        "dates": f"{start_formatted}/{end_formatted}",
        "details": description,
        "sf": "true"
    }
    return base_url + "&" + urllib.parse.urlencode(params)

def get_ics_content(title, date_str, time_str, description):
    """
    표준 .ics 파일 내용 생성
    """
    start_dt = datetime.strptime(f"{date_str} {time_str}", "%Y-%m-%d %H:%M")
    end_dt = start_dt + timedelta(minutes=30)
    
    start_formatted = start_dt.strftime("%Y%m%dT%H%M%S")
    end_formatted = end_dt.strftime("%Y%m%dT%H%M%S")
    
    lines = [
        "BEGIN:VCALENDAR",
        "VERSION:2.0",
        "PRODID:-//EasilySayAI//CalendarEvent//KO",
        "BEGIN:VEVENT",
        f"SUMMARY:{title}",
        f"DTSTART:{start_formatted}",
        f"DTEND:{end_formatted}",
        f"DESCRIPTION:{description}",
        "END:VEVENT",
        "END:VCALENDAR"
    ]
    return "\r\n".join(lines)

# 시뮬레이션 데이터 매핑
event_title = f"[{mock_tax_response['document_type_label']}] {mock_tax_response['important_dates'][0]['label']}"
event_date = mock_tax_response['important_dates'][0]['date']
event_time = mock_tax_response['important_dates'][0]['time']
event_desc = f"내용: {mock_tax_response['one_line_conclusion']}\n준비사항: {', '.join(mock_tax_response['required_items'])}"

print("=== 구글 캘린더 URL ===")
print(get_google_calendar_url(event_title, event_date, event_time, event_desc))
print("\n=== ICS 텍스트 본문 ===")
print(get_ics_content(event_title, event_date, event_time, event_desc))
````
